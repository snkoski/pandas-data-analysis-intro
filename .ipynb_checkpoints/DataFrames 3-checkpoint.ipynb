# Retrieve Rows by Index Label with .loc[ ]

```python
# use index_col to set index when making dataframe. use sort_index to sort dataframe by index (pandas works more efficiently if index is sorted)
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
# use .loc[] to find rows in dataframe. if only locating a single row it is returned as a Series
bond.loc["Goldfinger"]
```

```python
# bond.loc["Not A Bond Film"] .loc[] will return an error if index is not in dataframe
```

```python
# if duplicate indexes all rows will be returned
bond.loc["Casino Royale"]
```

```python
# can select a range of indexes
bond.loc["Diamonds Are Forever" : "Moonraker"]
```

```python
# select from chosen index to end of dataframe
bond.loc["GoldenEye" :]
```

```python
 # select from beginning of dataframe to chosen index
bond.loc[: "GoldenEye"]
```

```python
# can select multiple nonsequential rows
bond.loc[ ["GoldenEye", "Dr. No", "Thunderball"] ]
```

```python
# if index in list of indexes doesn't exist new row of Null values will be created
bond.loc[ ["Octopussy", "Moonraker", "Gold Bond"] ]
```

```python
# row of Nulls doesn't exist in original dataframe
"Gold Bond" in bond.index
```

# Second Arguments to .loc[ ] and .iloc[ ] 

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
# .loc[] can take a second argument that represents the column name you want to extract
bond.loc["Octopussy", "Actor"]
```

```python
# can extract multiple sequential columns
bond.loc["Moonraker", "Director" : "Budget"]
```

```python
# can use a list as second argument to get non-sequential columns
bond.loc["GoldenEye", ["Year", "Box Office", "Bond Actor Salary"]]
```

```python
# .iloc[] can also take second argument to extract column. second argument TAKES AN INT representing position of column in dataframe
# Actor is index position 1 in columns
bond.iloc[1, 1]
```

```python
# for sequential columns range is EXCLUSIVE, does not include second index
bond.iloc[14, 2 : 5]
```

```python
# can use a list to get non-sequential columns
bond.iloc[14, [2, 4, 1]]
```

```python
# can also get multiple columns from multiple rows
bond.iloc[[1,3,5], [5,3,1]]
```

# Set Multiple Values in DataFrame

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
# make conditional for column value you want to change
mask = bond["Actor"] == "Sean Connery"
```

```python
# use conditinal with .loc[] to find all rows with specified column value. this is NOT A COPY, it is a view of the rows from original dataframe
bond.loc[mask]
```

```python
# combine conditional with second argument to reassin values of column for each row
bond.loc[mask, "Actor"] = "Sir Sean Connery"
```

```python
# original dataframe has updated values
bond
```

# The .nsmallest() and .nlargest Methods

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
# can use .sort_values() along with column to sort by to get largest values in a column
bond.sort_values("Box Office", ascending=False).head(3)
```

```python
# can also use .nlargest() to get biggest value from specifed columns. first argument is number of rows to return, second argument is column to compare values
bond.nlargest(3, columns="Box Office")
```

```python
# use .nsmallest() to get rows with smallest values in specified column
bond.nsmallest(3, columns="Box Office")
```

```python
# can call .nlargest() on a Series, don't need to specify a column when you do this
bond["Box Office"].nlargest(5)
```

# Filtering with the .where() Method

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
# .where() takes a conditional and returns a dataframe with rows filled with Null where conditional is False 
# lets you get back original dataframe and visually represent which rows fit condition instead of returning a subset of original dataframe
mask = bond["Actor"] == "Sean Connery"
bond.where(mask).head()
```

```python
# can use multiple conditions
mask = bond["Actor"] == "Sean Connery"
mask2 = bond["Box Office"] > 800
bond.where(mask & mask2)
```

# The .query() Method

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
# .query() won't work on columns with spaces in their names. use Python .replace() to remove spaces
bond.columns = [column_name.replace(" ", "_") for column_name in bond.columns]
```

```python
bond.head(3)
```

```python
# column name and query value are entered as a string into .query()
# if searching for a string value must use second set of quotes
bond.query('Actor == "Sean Connery"')
```

```python
# order of quotes does not matter
bond.query("Director == 'Terence Young'")
```

```python
# for multiple conditions use 'and' or 'or' instead of '&' or '|'
bond.query("Actor == 'Roger Moore' and Director == 'John Glen'")
```

```python
# can select multiple values from single column with Python in statement and list of values
bond.query("Actor in ['Timothy Dalton', 'George Lazenby']")
```

```python
# can also use Python's not in statement
bond.query("Actor not in ['Sean Connery', 'Roger Moore']")
```

# A Review of the .apply() Method on Single Columns

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
# use .apply() to run functions over all values in a Series
def convert_to_string_and_add_millions(number):
    return str(number) + " MILLIONS!"

bond["Box Office"] = bond["Box Office"].apply(convert_to_string_and_add_millions)
```

```python
bond["Budget"] = bond["Budget"].apply(convert_to_string_and_add_millions)
```

```python
bond["Bond Actor Salary"] = bond["Bond Actor Salary"].apply(convert_to_string_and_add_millions)
```

```python
bond.head(3)
```

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
# can use Python for loop to perform .apply() on multiple columns
columns = ["Box Office", "Budget", "Bond Actor Salary"]
for col in columns:
    bond[col] = bond[col].apply(convert_to_string_and_add_millions)
```

```python
bond.head(3)
```

# The .apply() Method with Row Values

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
# write Python function that will take dataframe row as argument. saving values from row as variables and returning string bsed on values
# call .apply() with function as first argument and axis="columns" as second argument to apply function to each row
def good_movie(row):
    
    actor = row[1]
    budget = row[4]
    
    if actor == "Pierce Brosnan":
        return "The best"
    elif actor == "Roger Moore" and budget > 40:
        return "Enjoyable"
    else:
        return "I have no clue"
    
bond.apply(good_movie, axis="columns").head()
```

# The .copy() Method

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
# assign values of a column to a variable
directors = bond["Director"]
directors.head(3)
```

```python
# get warning if you try to change the values of the Series in your new variable
directors["A View to a Kill"] = "Mister John Glen"
```

```python
# value in Series has been changed
directors.head(3)
```

```python
# original dataframe has also been changed
bond.head(3)
```

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
# .copy() creates a copy of a dataframe or Series but stores it separately in memory
directors = bond["Director"].copy()
directors.head(3)
```

```python
# change value in .copy() variable
directors["A View to a Kill"] = "Mister John Glen"
```

```python
# value is changed
directors.head(3)
```

```python
# can change values in .copy() variable withour affecting the original dataframe
bond.head(3)
```
